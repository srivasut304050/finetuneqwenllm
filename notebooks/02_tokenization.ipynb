# Lesson 3: Tokenization & Attention Masks

Welcome to **Lesson 3**! In this notebook, we explore the tokenization process using the Qwen model's tokenizer. We convert characters into lists of token indices, set up padding/truncation configurations, and inspect attention masks.

### 1. Load Tokenizer

```python
import sys
sys.path.append('..')
from transformers import AutoTokenizer
from src.tokenizer import load_qwen_tokenizer

model_id = "Qwen/Qwen2.5-1.5B-Instruct"
tokenizer = load_qwen_tokenizer(model_id)
```

### 2. Tokenize a Simple Phrase

```python
text = "Fine-tuning Qwen models with LoRA!"

# Convert text to token IDs
tokens = tokenizer.tokenize(text)
ids = tokenizer.encode(text)

print(f"Raw Text:  {text}")
print(f"Tokens:    {tokens}")
print(f"Token IDs: {ids}")
```

### 3. Special Tokens & Decoding

```python
# Inspect vocabulary details
print(f"Vocab Size: {tokenizer.vocab_size}")
print(f"EOS Token:  {tokenizer.eos_token} (ID: {tokenizer.eos_token_id})")
print(f"PAD Token:  {tokenizer.pad_token} (ID: {tokenizer.pad_token_id})")

# Decode token IDs back to text
decoded = tokenizer.decode(ids)
print(f"Decoded:    {decoded}")
```

### 4. Attention Masks and Padding

```python
sentences = [
    "Short sentence.",
    "This is a significantly longer sentence to demonstrate padding mechanisms."
]

# Tokenize with padding
inputs = tokenizer(
    sentences,
    padding=True,
    truncation=True,
    max_length=32,
    return_tensors="pt"
)

print("Input IDs (Tokens padded to match size):")
print(inputs['input_ids'])
print("\nAttention Mask (0s for padded indices, 1s for real characters):")
print(inputs['attention_mask'])
```
