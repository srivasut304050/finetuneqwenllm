# Lesson 2: Dataset Exploration & Preprocessing

Welcome to **Lesson 2**! In this notebook, we will explore conversational datasets, structure them into standard instruction formats, and calculate basic dataset properties like text length and token estimates.

### 1. Import Dependencies

```python
import json
import os
import pandas as pd
import matplotlib.pyplot as plt
```

### 2. Load the Raw Data

```python
# Load our sample instruction set generated during setup
raw_data_path = '../data/raw/sample.json'

with open(raw_data_path, 'r') as f:
    dataset = json.load(f)

print(f"Number of dialog samples: {len(dataset)}")
```

### 3. Display Conversations

```python
# Print a structured dialog sample
for i, sample in enumerate(dataset):
    print(f"\n--- Sample {i+1} ---")
    for msg in sample['messages']:
        print(f"[{msg['role'].upper()}]: {msg['content']}")
```

### 4. Sequence Length Analysis

```python
# Measure text length distribution
lengths = []
for sample in dataset:
    total_chars = sum(len(msg['content']) for msg in sample['messages'])
    lengths.append(total_chars)

df = pd.DataFrame(lengths, columns=['character_length'])
print(df.describe())
```
